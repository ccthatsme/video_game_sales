## Altair Charts


```python
import pandas as pd
import altair as alt
import numpy as np
```

```python
df = pd.read_csv('../data/cleaned_data/vgchartz-2024-cleaned.csv')
df.head()
```

```python
sales_by_region = df.groupby('year')[['na_sales', 'pal_sales', 'jp_sales', 'other_sales']].sum().reset_index()

melted_sales_region = sales_by_region.melt(value_vars=['na_sales', 'pal_sales', 'jp_sales', 'other_sales'], value_name="sales", var_name="region", id_vars="year", ignore_index=False)

```

```python
alt.Chart(melted_sales_region).mark_line().encode(
    x='year:O',
    y='sales:Q',
    color='region:N').properties(
        width=500,
        height=200
    )
```

## Supervised Learning Tasks

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../data/cleaned_data/vgchartz-2024-cleaned.csv')
df.head(20)
```

```python
df_rockstar = df
df_rockstar['developer'].nunique()

df_rockstar['year'].isna().sum()
```

```python
## remove where we dont have a critic_score
df_rockstar = df_rockstar[df_rockstar['critic_score'].notna()]
df_rockstar = df_rockstar.drop(columns=['img', 'release_date', 'last_update', 'title', 'developer', 'publisher', 'na_sales', 'jp_sales', 'pal_sales', 'other_sales',])
df_rockstar.isna().sum()
df_rockstar = df_rockstar.dropna(subset=['year'])
df_rock = pd.get_dummies(df_rockstar, columns=['console', 'genre'], drop_first=True)
# df_rock = df_rock.drop(columns=['total_sales'])
df['year'].unique()[:20]

```

```python
df_rock['year'].isna().sum()
```

## Train test splitting

```python
X_train, X_test, y_train, y_test = train_test_split(df_rock.drop(columns=['total_sales']), df_rock['total_sales'], test_size=0.2, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## training linear classifier

```python
model = LinearRegression().fit(X_train, y_train)
```

```python
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
print(model.intercept_)
print(model.coef_)
print(df_rock.columns)
```
